```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names = ['Location', 'Barcode', 'SKU', 'Item Name', 'Q', 'Unit']
```

```python
df = pd.read_csv('092118-02.csv', header = None)
df = df.replace(r'\n','', regex=True)
df.columns = column_names
```

```python
# DRW为string，Loc为int
df['DRW'] = df['Location'].str.split('-').str[2]
df['Loc'] = df['DRW'].str.extract('(\d+)')
df['Loc'] = df['Loc'].fillna('1').astype(int)
```

```python
# DRW为string，Loc为int
df['DRW'] = df['Location'].str.split('-').str[2]
df['Loc'] = df['DRW'].str.extract('(\d+)')
df['Loc'] = df['Loc'].fillna('1').astype(int)
```

```python
# 检索每一行，如果df['DRW'] 等于 DR5，将那一行的df['Loc']变为3
df.loc[df['DRW'].str.contains('DR5'), 'Loc'] = 3
```

```python
df.loc[df['DRW'].str.contains('REF'), 'Loc'] = 181
```

```python
# 把 Num从list变为数字（因为有1B-1）
df['NUM'] = df['Location'].str.split('-').str[3:]
df['NUM1'] = [','.join(map(str, l)) for l in df['NUM']]
df.NUM1 = pd.to_numeric(df.NUM1, errors='coerce')
df
```

```python
# 给ref设置一个num1，这样，他就不会被设置成ladies里
df.loc[df['DRW'].str.contains('REF'), 'NUM1'] = 200
```

```python
df.head(20)
```

```python
ladies = df[(df['NUM1'] <= 36) | (df['NUM1'].isnull())]
ladies
```

```python
#添加映射关系来排序字母
mapping_dict = {
    'A': 1000,
    'B': 2000,
    'C': 3000,
    'D': 4000,
    'F': 5000,
    'G': 6000,
    'H': 7000,
    'I': 8000,
    'J': 9000,
    'K': 10000,
    'L': 11000
}
ladies['NUM1'] = ladies['NUM1'].fillna(ladies['Location'].str.split('-').str[-2].str[0].map(mapping_dict))
```

```python
ladies = ladies.sort_values(['Loc', 'NUM1'], ascending = [True, True])
ladies
```

```python
ladies = ladies.drop(['NUM', 'NUM1', 'DRW', 'Loc'], axis = 1)
```
